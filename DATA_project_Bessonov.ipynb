```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pds
file =('Startups1.csv')
df = pds.read_csv(file)
```

Содержание: Следующий набор данных содержит данные о 300 лучших стартапах Индии.
Подробная информация о столбцах:

(Company)Компания - Название стартапа.

(City)Город - Город, в котором запущен стартап.

(Starting Year)Год начала - Год, в котором был запущен стартап.

(Founders)Учредители - Имя основателей стартапа.

(Industries)Отрасли - отраслевая область, к которой относится стартап.

(No. of Employees)Количество сотрудников - Количество сотрудников в стартапе.

(Funding Amount in USD)Сумма финансирования в долларах США - Общая сумма финансирования, выделенная стартапу.

(Funding Rounds)Раунды финансирования - это количество раз, когда стартап возвращается на рынок, чтобы привлечь больше капитала. Цель каждого раунда состоит в том, чтобы основатели обменяли акции своего бизнеса на капитал, который они могут использовать для продвижения своих компаний на следующий уровень.

(No. of Investors)Количество инвесторов - Количество инвесторов в стартапе.

 3)В каком городе больше всего открылось стартапов?

```python
cities = df['City']
city_counts = cities.value_counts()
fig, ax = plt.subplots(figsize = (15, 15))
ax.pie(city_counts.values, labels=city_counts.index, autopct='%1.1f%%')
ax.set_title('Распределение стартапов по городам', size=15)
plt.show()
```

Вывод: Bengaluru действительно оказался самым популярным городом для стартапов, при этом столица Индии New Dellhi оказалась лишь на 4-ом месте.

4) В каком году открылось больше всего новых стартапов?

```python
plt.subplots(figsize=(16, 3))
df['Starting Year'].value_counts().plot(kind='bar')
plt.title("Количество стартапов по годам", fontsize=15)
plt.xlabel("Год", fontsize=15, color="g")
plt.ylabel("Количество стартапов", fontsize=15, color="g")
plt.show()
```

Вывод: в 2015 году было открыто больше всего индийских стартапов

5) Найдем компанию с наибольшим количеством раундов финансирования

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Funding Round',y='Company',data=df.sort_values('Funding Round', ascending=False).head(10))
plt.title('Компании с наибольшим количеством раундов финансирования',fontsize=20)
plt.ylabel('Компании',fontsize=20)
plt.xlabel('Количество раундов',fontsize=20)
plt.show()
```

Вывод: наибольшее количество раундов финансирования у компании Ola, как наибольшее количество инвесторов

6) Какие компании, получили максимальную сумму финансирования в $?

```python
df.groupby(["Starting Year","Company","City"])["Funding Amount in $"].max().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x='Funding Amount in $',y='Company',data=df.sort_values('Funding Amount in $', ascending=False).head(10))
plt.title('Компании с самой большой суммой финансирования',fontsize=20)
plt.ylabel('Компании',fontsize=20)
plt.xlabel('Сумма финансирования $',fontsize=20)
plt.show()
```

Вывод: наибольшее финонсирование привлекла компания Reliance Jio/, а Ola c наибольшим количество раундов финансирования и c наибольшим количеством инвесторов только на 5-ом месте

7) Какое самое популярное количество работников в стартапах?

```python
plt.figure(figsize=(10,10))
plt.title('Количество работников',fontsize=20)
abs_values = df['No. of Employees'].value_counts().values
ax = sns.countplot(data=df, y=df['No. of Employees'],order=df['No. of Employees'].value_counts().index)
ax.bar_label(container=ax.containers[0],labels=abs_values)
plt.show()
```
